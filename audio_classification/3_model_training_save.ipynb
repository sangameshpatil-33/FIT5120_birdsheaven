```python
import pandas as pd
import numpy as np
```

```python
extracted_features_df = pd.read_csv('feature_vector_specie_df.csv')
extracted_features_df.head(10)
```

```python
for i in range(len(extracted_features_df)):
    extracted_features_df.feature[i] = extracted_features_df.feature[i].strip('[')
    extracted_features_df.feature[i] = extracted_features_df.feature[i].strip(']')
    extracted_features_df.feature[i] = extracted_features_df.feature[i].strip('\n')
#     if '' in extracted_features_df.feature[i]:
#         extracted_features_df.feature[i] = extracted_features_df.feature[i].split('')
#     while("" in extracted_features_df.feature[i]) :
#         extracted_features_df.feature[i].remove("")
```

```python
float(extracted_features_df.feature[2].split(",")[0])
```

```python
features = []
for i in extracted_features_df.feature:
    m = i.split(",")
    lst =[]
    for j in m:
        lst.append(float(j))
    print(lst)
    features.append(lst)
        
extracted_features_df.feature = features
```

```python
extracted_features_df['feature']
```

```python
X=np.array(extracted_features_df['feature'].tolist())
y=np.array(extracted_features_df['specie'].tolist())
```

```python
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
y= to_categorical(labelencoder.fit_transform(y))
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
len(X_train), len(X_test)
```

```python
import tensorflow as tf
print(tf.__version__)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout,Activation,Flatten
from tensorflow.keras.optimizers import Adam
from sklearn import metrics
```

```python
num_labels=4
```

```python
from keras.models import Sequential
from keras.layers import Activation, Dense, Dropout

model=Sequential()
###first layer
model.add(Dense(100,input_shape=(40,)))
model.add(Activation('relu'))
model.add(Dropout(0.5))
###second layer
# model.add(Dense(200))
# model.add(Activation('relu'))
# model.add(Dropout(0.5))
###third layer
# model.add(Dense(100))
# model.add(Activation('relu'))
# model.add(Dropout(0.5))

###final layer
model.add(Dense(num_labels))
model.add(Activation('softmax'))
```

```python
model.compile(loss='categorical_crossentropy',metrics=['accuracy'],optimizer='adam')
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint
from datetime import datetime 

num_epochs = 200
num_batch_size = 32

checkpointer = ModelCheckpoint(filepath='saved_models/', verbose=1, save_best_only=True)

start = datetime.now()

model.fit(X_train, y_train, batch_size=num_batch_size, epochs=num_epochs, validation_data=(X_test, y_test), verbose=1)


duration = datetime.now() - start
print("Training completed in time: ", duration)
```

```python
test_accuracy = model.evaluate(X_test,y_test,verbose=0)
print(test_accuracy[1])
```
